```python
import matplotlib.pyplot as plt
import numpy as np
import sys
from scipy.optimize import linprog
```

### Definição da função problema

```python
def max_profit(x1, x2, x3) -> float:
    return 33*x1 + 12*x2 + 19*x3
```

### Resolução do problema

```python
section = np.array(['Frezadoras', 'Tornos','Furadoras'])
objective = np.array([-33, -12, -19])
availability = np.array([500, 350, 150, 20])
prod_need = np.array([
  [9, 5, 3],
  [3, 4, 0],
  [5, 0, 2],
  [0, 1, 0]
])
bounds = [(0, None), (0, None), (0, None)]

result = linprog(c=objective, A_ub=prod_need, b_ub=availability, bounds=bounds)

if not result.success:
  print(f"Não foi possível resolver esse problema.")
  sys.exit(1)

best_prod = result.x
best_profit = (objective * (-1)) * best_prod
print(f"""A melhor produção é {best_prod[0]} P1, {best_prod[1]} P2 e {best_prod[2]} P3,
      que gera um lucro máximo de {np.sum(best_profit)}""")
```
